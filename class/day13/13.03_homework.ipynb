```python
from selenium import webdriver 
from bs4 import BeautifulSoup
driver = webdriver.Chrome('C:/tutorial/chromedriver.exe')
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2017-01-01,2017-12-31&count=100&start=1&ref_=adv_nxt")
driver.implicitly_wait(3)
```

```python
#rank, title, description, runtime, genre, rating, metascore, votes, gross_earning_in_Mil, director, actor
#순위,제목,설명,상영시간,장르,평가,매타스코어,투표,검독,총수입,배우
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
import pandas as pd
col1 = {'rating':[1,2,3]}

col1['name']=["hyejin","hi","m"]
df = pd.DataFrame(col1)
print(df)
```

```python
dic ={}
```

```python
#rating
#평점가져오기
arr=soup.find_all('div' , attrs={'class':'inline-block ratings-imdb-rating'})
ratings=[]
ranks= []
for num,i in enumerate(arr,start =1) :
    ratings.append(i.strong.text)
    ranks.append(num)
dic["ratings"] =ratings
dic["ranks"]=ranks
dic
```

```python
#title
#이름 가져오기 
#main > div > div.lister.list.detail.sub-list > div > div:nth-child(1) > div.lister-item-content > h3 > a
# h3.lister-item-header>a
arr=soup.select("h3.lister-item-header>a")
titles = [i.text.strip() for i in arr]
dic['titles'] =titles
```

```python
#description
arr=soup.select("div.lister-item-content >p:nth-child(4)")
descriptions = [i.text.strip() for i in arr]
```

```python
#runtime
#런타임
arr=soup.select("div.lister-item-content >p:nth-child(2)>span.runtime")
runtimes = [int(i.text.replace('min','').strip()) for i in arr]
dic['runtimes'] =runtimes
```

```python
#genre
#장르
# 스크래핑할 데이터 - rank, title, description, runtime, genre, rating, metascore, votes, gross_earning_in_Mil, director, actor
infos=soup.select("div.lister-item-content >p:nth-child(2)>span.genre")
genres=[info.text.strip() for info in infos ]
dic['genres'] =genres
df = pd.DataFrame(dic)
df
```

```python
import numpy as np
#metascore
##main > div > div.lister.list.detail.sub-list > div > div:nth-child(1) > div.lister-item-content > div > div.inline-block.ratings-metascore > span

arr = soup.select("div.inline-block.ratings-metascore > span")
metascore = [str(k)+"= "+i.text.strip() if i is not None else '없음'for k,i in enumerate(arr)  ]
print(metascore)
print(len(metascore))

```

```python
import numpy as np
#metascore
##main > div > div.lister.list.detail.sub-list > div > div:nth-child(1) > div.lister-item-content > div > div.inline-block.ratings-metascore > span
#main > div > div.lister.list.detail.sub-list > div > div:nth-child(19) > div.lister-item-content > div
arr = soup.select("div.lister-item-content > div")


metascores =[]
for i in arr:
    tmp=i.select_one("div.inline-block.ratings-metascore > span")
    if tmp is None :
        metascores.append("없음")
    else:
        metascores.append(tmp.text.strip())
metascores
len(metascores)
```

```python
#director,actor
#감독 배우 찾기
infos = soup.select("div.lister-item-content > p:nth-child(5)")
directors =[]
stars =[]
for info in infos:
    list= info.text.split()
    del list[0]
    #print(i)
    directorstr=''
    starslist =[]
    start=0
    for i, name in enumerate(list):
        if name=='|':
            start=i+2
            break
        directorstr +=", "+name
    starstr=''
    for i in list[start:]:
        starstr += ' '+i
    #print(director)
    #print(star)
    directors.append(directorstr)
    stars.append(starstr)
print(directors)
print(stars)
```

```python
#votes
#투표
infos=soup.select("p.sort-num_votes-visible > span:nth-child(2)")
votes =  [int(info.text.replace(',','').strip()) for info in infos ]
```

```python
#gross_earning_in_Mil
#총수입
infos=soup.select("p.sort-num_votes-visible > span:nth-child(5)")
gross_earning_in_Mils =\
[int(info.text.replace('$','').replace('M','').replace('.','')) for info in infos]
gross_earning_in_Mils
```
