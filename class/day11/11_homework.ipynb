```python
# 데이터 읽어오기 
import pandas as pd
import numpy as np
```

```python
excel_dir = './2019년_11월_교통카드_통계자료.xls'
df= pd.read_excel(excel_dir, sheet_name = '지하철 시간대별 이용현황', header=None)
df.head()
```

```python
#작업일시 뽑기
work_date = df.iloc[2,52]
print(work_date)
print(type(work_date))
df=df.drop([0,2,52],axis=1)

df
```

```python
# 시간데이터 가져오기
ser =df.iloc[0,:]
ser=ser.dropna(axis =0)
ser = ser.drop([1,3])
ser
```

```python
import matplotlib.pyplot as plt
#matplotlib 한글 폰트 오류 문제 해결

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

plt.figure(figsize=(100, 50))
```

```python
# 데이터 바꾸기
n,m = df.shape
npd=pd.DataFrame()
plt.figure(figsize=(50, 30))
for i in range(2,n):
    x1_values = df.iloc[i,2::2]
    x1_values=x1_values.str.replace(pat=',', repl='', regex=False)
    x1_values = pd.to_numeric(x1_values)
    nser = pd.Series(x1_values.values ,index =ser)
    npd[df.iloc[i,1]] =nser
npd.info()
npd
```

```python
npd.plot(kind='bar', width=0.7,  figsize=(50, 20))
```

```python
# 출근시간대 7~9시에 가장 많이 타고 내리는 역 찾기
pr2=npd.loc[["19:00:00~19:59:59","20:00:00~20:59:59"],:]
pr2
```

```python
ser=pr2.sum(axis = 0 )
m=ser.max()
ser[ser ==m]
```

```python
# 출근시간대 7~9시에 가장 많이 타고 내리는 역 찾기
pr3=npd.loc["23:00:00~23:59:59",:]
m=pr3.max(axis =0)
m
pr3[pr3 ==m]
```
