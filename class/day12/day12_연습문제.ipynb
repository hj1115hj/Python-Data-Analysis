 박스오피스 1~5위 영화제목하고 평점, 댓글 내용을 10개씩만 추출하여 출력하는 웹크로링 소스를 구현 단 평점이 있더라도 댓글이 비어있는 것은 제외

```python
from bs4 import BeautifulSoup
import requests
import re
# 영화 순위 페이지 읽어오기
req = requests.get('https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&q=%EC%98%81%ED%99%94')
html =req.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 영화 5개 목록의 id 추출
test=soup.select_one("#morColl > div.coll_cont > div > ol")

movies_id =[]
for i in range(1,6):
    href=test.select_one('li:nth-child(%d) > div.wrap_cont.cont_type2 > div > a'%i).attrs['href']
    movie_id=re.findall("\d{3,}",href)
    movies_id.extend(movie_id)
```

```python
#영화 제목, 평점, 댓글 10개 추출 비어있는 댓글은 평점이 있더라도 제외
for movie_id in movies_id:
    url ='https://movie.daum.net/moviedb/grade?movieId='+movie_id
    req = requests.get(url)
    html =req.text
    soup = BeautifulSoup(html, 'html.parser')
    name=soup.select_one("#mArticle > div.view_movie > a > h2")
    name = name.text
    print("영화이름: "+name)
    scores = soup.find_all("em", class_ ="emph_grade")
    reviews =soup.find_all("p",class_='desc_review')
    for i,review in enumerate(reviews) :
        if review.text.strip() == '':
            continue
        print(i+1,'번 리뷰')
        print("평점:",scores[i].text.strip(), end=' ')
        print(review.text.strip())    
      
```
